```python
#merge three or more than three dfs
import pandas as pd
df1 = pd.DataFrame({"A":[1,2,3],"B":[4,5,6]})
df2 = pd.DataFrame({"C":[7,8,9],"D":[10,11,12]})
df3 = pd.DataFrame({"E":[13,14,15],"F":[16,17,18]})
pd.merge(df1,df2,df3)
#result
#merge can only merge two data frames ,for three we have to use reduce or other functions
```

```python
#Perform join on df and series
df1 = pd.DataFrame({"A":[1,2,3],"B":[4,5,6]})
df2 = pd.DataFrame({"C":[7,8,9],"D":[10,11,12]})
df1.join(df2)
sr1 = pd.Series([1,2,3],name = "A")
sr2 = pd.Series([4,5,6], name = "B")
#series does not support join , it only supports concat we can join it using different methods
sr1.to_frame().join(sr2)
```

```python
#join on concat
sr1 = pd.Series([1,2,3],name = "A")
sr2 = pd.Series([4,5,6], name = "B")
result =pd.concat([sr1,sr2],axis=1)
sr3 = pd.Series([7,8,9],name = "C")
result.join(sr3)
```
