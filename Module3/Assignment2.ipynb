```python
#Mention the data point names with the help of loops
import matplotlib.pyplot as plt


x = ["Aayush","Mayank","Ram","Mohan","Shyam"]
y = [10, 20, 25, 30, 40]
plt.plot(x,y,color="r", marker="o")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Students vs Marks")
plt.scatter(x, y, color="b")
for i in range(len(x)):
    plt.text(i + 0.1, y[i] + 0.1, x[i], fontsize=10)
plt.show()
```

Can we draw two lines in a single graph?
Yes, we can draw

```python
z = [40,90,80,70,60]
plt.plot(x,y,color ="r", marker="o")
plt.plot(x,z,color="b",marker="o")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Students vs Marks")
plt.scatter(x, y, color="b")
plt.show()
```

```python
#Add legend in your graph
z = [40,90,80,70,60]
plt.plot(x,y,color ="r", marker="o",label="Maths")
plt.plot(x,z,color="b",marker="o",label="Science")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Students vs Marks")
plt.scatter(x, y, color="b")
plt.legend()
plt.show()
```

```python
#draw subplots for the line chart

plt.plot(x,y,color ="r", marker="o",label="Maths")
plt.plot(x,z,color="b",marker="o",label="Science")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Students vs Marks")

fig, axs = plt.subplots(1, 2, figsize=(12, 5))  # axs[0] and axs[1]

# First subplot: Math marks
axs[0].plot(x, y, color="r", marker="o", label="Math")
axs[0].set_title("Math Marks")
axs[0].set_xlabel("Students")
axs[0].set_ylabel("Marks")
axs[0].legend()

# Second subplot: Science marks
axs[1].plot(x, z, color="b", marker="s", label="Science")
axs[1].set_title("Science Marks")
axs[1].set_xlabel("Students")
axs[1].set_ylabel("Marks")
axs[1].legend()

# Adjust spacing between subplots
plt.tight_layout()
plt.show()
```
