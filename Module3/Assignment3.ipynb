```python
#libraries
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv('penguins.csv')
df.head()
```

**Scatter plot in seaborn**

```python
sns.scatterplot(data=df, x='bill_length_mm', y='bill_depth_mm', hue='species')
plt.figure(figsize=(10,10))
sns.set_theme(style='darkgrid',context="notebook")
plt.show()
```

**line plot**

```python
sns.lineplot(x='bill_length_mm', y='bill_depth_mm', data=df, hue='species')
plt.figure(figsize=(10,10))
sns.set_style('darkgrid')
plt.show()
```

**Histogram Plot**

```python
sns.histplot(data=df, x='bill_length_mm', hue='species')
plt.figure(figsize=(10,10))
sns.set_style('darkgrid')
plt.show()
```

**Box Plot**

```python
sns.boxplot(data=df, x='species', y='bill_length_mm')
plt.figure(figsize=(10,10))
plt.show()
```

**Violin plot**

```python
sns.violinplot(data=df, x='species', y='bill_length_mm')
plt.figure(figsize=(10,10))
plt.show()
```

**Bar plot**

```python
sns.barplot(data=df, x='species', y='bill_length_mm')
plt.figure(figsize=(10,10))
plt.show()
```

**Count plot**

```python
sns.countplot(data=df, x='species')
plt.figure(figsize=(10,10))
plt.show()
```
